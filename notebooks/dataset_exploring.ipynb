```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
roco_train_path = "/Users/pdcos/Documents/Mestrado/IA025/Projeto_Final/Code/caption_medical_images/dataset_structure/roco-dataset/data/train/radiology/captions.txt"
```

```python
df_train = pd.read_csv(roco_train_path, sep="\t", header=None)
rename_dict = {0: "IMG_NAME", 1:"CAPTIONS"}
df_train.rename(columns=rename_dict, inplace=True)
df_train.head()
```

```python
#df_train['TOTAL_WORDS'] = df_train['IMG_NAME'].str.count('') + 1
df_train['TOTAL_WORDS'] = [len(x.split()) for x in df_train['CAPTIONS'].tolist()]
```

```python
ax = df_train.plot.hist(column=["TOTAL_WORDS"], figsize=(20, 10), bins=400, xlim=(0,128))
```
